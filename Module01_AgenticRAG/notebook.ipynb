```python
import requests

url = "https://datatalks.club/faq/json/courses.json"
response = requests.get(url)
curses_raw = response.json()
```

```python
curses_raw
```

```python
documents = []
prefix = 'https://datatalks.club/faq'
for course in curses_raw:
    course_url = f'{prefix}{course["path"]}'

    course_response = requests.get(course_url)
    course_response.raise_for_status()
    curses_raw = course_response.json()

    documents.extend(curses_raw)
len(documents)
```

```python
documents[0]
```

```python
from minsearch import Index

index = Index(
    text_fields=["section", "question", "answer"],
    keyword_fields=['course']
)

index.fit(documents)
```

```python
def search(query: str, course: str):
    return index.search(
        query=query,
        filter_dict={'course': course},
        num_results=5
    )
```

```python
def rag_pipeline(query: str, course: str, model):
    context = search(query, course)
    query = prompt_builder(query, context)
    answer = model(query)
    print(answer)
```

```python
def prompt_builder(query, context):
    return f"""
    Instructions: You are AI assistant your task is to help and answer the user's question's only using the following context
    ---
    question: {query}
    ---
    context: {context}
    ---
    give the direct answer
    """
```

```python
def build_context(search_result):
    lines = []
    for docs in search_result:
        lines.append(docs['section'])
        lines.append("Q:" + docs['question'])
        lines.append("A:" + docs['answer'])
        lines.append('')
    return "\n".join(lines).strip()
```

```python
def rag_pipeline(query: str, course: str, model, search):
    search_result = search(query, course)
    context = build_context(search_result)
    query = prompt_builder(query, context)
    answer = model(query)
    print(answer)
```

```python
search_result = search(query, "llm-zoomcamp")
context = build_context(search_result)
print(context)
```

```python
rag_pipeline(query, course="llm-zoomcamp", model=llm, search=search)
```

```python
def system_message(context):
    return f"""Instructions: You are AI assistant
                     task your task is to help and answer the user's question's only using the following context: {context})
                     give the direct answer"""
```

```python
def prompt_builder(query, system_message):
    return [
        {'role': "system", "content": system_message},
        {'role': "user", "content": query}
    ]
```

```python
def rag_pipeline(query: str, course: str, model, search):
    search_result = search(query, course)
    context = build_context(search_result)
    system_message_ = system_message(context)
    query = prompt_builder(query, system_message_)
    answer = model(query)
    print(answer)
```

```python
rag_pipeline(query, course="llm-zoomcamp", model=llm, search=search)
```
