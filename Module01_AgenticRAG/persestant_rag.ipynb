```python
from rag_helper import RagBase
from openai import OpenAI
import os
from sqlitesearch import TextSearchIndex
from dotenv import load_dotenv
```

```python
load_dotenv()
openai_clint = OpenAI(
    api_key=os.getenv("API_KEY"),
    base_url=os.getenv("API_URL"),
)
```

```python
index = TextSearchIndex(
    text_fields=["section", "question", "answer"],
    keyword_fields=['course'],
    db_path='faq.db'
)
```

```python
len(index.search('i just discover the course can i join now?'))
```

```python
assistant = RagBase(index, openai_clint)
```

```python
query = "i just discover the course can i join now?"
assistant.rag_pipeline(query)
```
